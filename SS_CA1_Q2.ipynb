```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 20
```

```python
def imp(n):
  if n==0:
    return 1
  return 0

def u(n):
  if n>=0:
    return 1
  return 0
```

# A.

```python
h = []
for i in range(-int(N/2), int(N/2)+1):
  h.append((imp(i)+imp(i-1))/2)
plt.stem(np.arange(-int(N/2), int(N/2)+1), h, use_line_collection=True)
plt.grid(True)
plt.show()
```

```python
S = []
for i in range(-int(N/2), int(N/2)+1):
  S.append((u(i)+u(i-1))/2)
plt.stem(np.arange(-int(N/2), int(N/2)+1), S, use_line_collection=True)
plt.grid(True)
plt.show()
```

# B.

```python
h = []
for i in range(-int(N/2), int(N/2)+1):
  h.append(imp(i+1)-imp(i))
plt.stem(np.arange(-int(N/2), int(N/2)+1), h, use_line_collection=True)
plt.grid(True)
plt.show()
```

```python
S = []
for i in range(-int(N/2), int(N/2)+1):
  S.append(u(i+1)-u(i))
plt.stem(np.arange(-int(N/2), int(N/2)+1), S, use_line_collection=True)
plt.grid(True)
plt.show()
```

# C.

```python
h = []
for i in range(-int(N/2), int(N/2)+1):
  h.append(imp(i+1)-2*imp(i)+imp(i-1))
plt.stem(np.arange(-int(N/2), int(N/2)+1), h, use_line_collection=True)
plt.grid(True)
plt.show()
```

```python
S = []
for i in range(-int(N/2), int(N/2)+1):
  S.append(u(i+1)-2*u(i)+u(i-1))
plt.stem(np.arange(-int(N/2), int(N/2)+1), S, use_line_collection=True)
plt.grid(True)
plt.show()
```

```python
h = []
for i in range(-int(N/2), int(N/2)+1):
  temp = 0
  for m in range(0, 20):
    temp += np.cos((np.pi * i)/10) * imp(i-m)
  h.append(temp)
plt.stem(np.arange(-int(N/2), int(N/2)+1), h, use_line_collection=True)
plt.grid(True)
plt.show()
```
