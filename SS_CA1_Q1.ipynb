## Question 1

```python
import numpy as np
from numpy import pi
```

```python
import matplotlib.pyplot as plt
```

```python
N = 25
n = np.arange(-(N/2), N/2 + 1, 1)
k = .0625
T = (2*pi)/(2*k*pi)
```

```python
x = 2 * np.cos(2 * pi * k * n)
plt.stem(n, x, use_line_collection=True) 
plt.title('x[n] = 2cos(2kpin)')
plt.show()
```

```python
yA = []
for i in range(0, N):
  yA.append((x[i + 1] - x[i])/T)

plt.stem(np.arange(-(N/2), N/2, 1), yA, use_line_collection=True) 
plt.title('A. y[n] = 1/T(x[n+1]-x[n])')
plt.show()
```

```python
yB = []
for i in range(1, N):
  yB.append((x[i + 1] - x[i - 1])/(2*T))

plt.stem(np.arange(-(N/2) + 1, N/2, 1), yB, use_line_collection=True) 
plt.title('B. y[n] = 1/2T(x[n+1]-x[n-1])')
plt.show()
```

```python
yC = []
for i in range(1, N):
  yC.append((x[i + 1] - 2*x[i] + x[i - 1])/(T**2))

plt.stem(np.arange(-(N/2) + 1, N/2, 1), yC, use_line_collection=True) 
plt.title('C. y[n] = 1/T**2(x[n+1]-2x[n]+x[n-1])')
plt.show()
```

```python
yD = []
for i in range(2, N-1):
  yD.append(0.1*x[i + 2] + 0.2*x[i + 1] + 0.4*x[i] + 0.1*x[i - 2] + 0.2*x[i - 1])

plt.stem(np.arange(-(N/2) + 2, (N/2) - 1, 1), yD, use_line_collection=True) 
plt.title('D. y[n] = 0.1x[n-2]+0.2x[n-1]+0.4x[n]+0.2x[n+1]+0.1x[n+2])')
plt.show()
```
