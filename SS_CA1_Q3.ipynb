```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 20
```

```python
def imp(n):
  if n==0:
    return 1
  return 0

def u(n):
  if n>=0:
    return 1
  return 0
```

```python
y0_4 = []
for i in range(-int(N/2), int(N/2)+1):
  if i>=0 and i<=4:
    y0_4.append(2-(1/(2**i)))
  else:
    y0_4.append(0)
plt.stem(np.arange(-int(N/2), int(N/2)+1), y0_4, use_line_collection=True)
plt.grid(True)
plt.title('y[n] for 0<n<4')
plt.show()
```

```python
y4_ = []
for i in range(-int(N/2), int(N/2)+1):
  if i<=4:
    y4_.append(0)
  else:
    y4_.append((1/(2**(i-5)))-(1/(2**i)))
plt.stem(np.arange(-int(N/2), int(N/2)+1), y4_, use_line_collection=True)
plt.grid(True)
plt.title('y[n] for 4<n')
plt.show()
```

```python
y = []
j = 0
for i in range(-int(N/2), int(N/2)+1):
  y.append(y0_4[j] + y4_[j])
  j+=1
plt.stem(np.arange(-int(N/2), int(N/2)+1), y, use_line_collection=True)
plt.grid(True)
plt.title('y[n]')
plt.show()
```
